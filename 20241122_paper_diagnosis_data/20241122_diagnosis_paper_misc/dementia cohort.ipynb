```r
#codelist
CREATE TABLE `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.cb_codelist_dementia_all_cause` as  
SELECT *,CAST(SNOMEDCT_CONCEPTID AS STRING) AS snomedcode 
FROM `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.cb_codelist_dementia_allcause`

#20240418
CREATE TABLE `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.pj1_dementia_all` AS
SELECT person_id,  min(tbl_srcode_start_date) dementia_variable_start_time, 1 AS dementia_variable_value
from `yhcr-prd-phm-bia-core.CB_FDM_PrimaryCare_V9.tbl_srcode` 
WHERE snomedcode in 
(SELECT snomedcode from `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.cb_codelist_dementia_allcause`)  
GROUP BY person_id

CREATE TABLE `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.pj1_dementia_13-19`
AS SELECT * from `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.pj1_dementia_all`
WHERE(dementia_variable_start_time>='2013-01-01' AND dementia_variable_start_time<='2019-12-31') 

# expand the table
Create table `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.dementia_3` as
Select distinct a.person_id
  , b.dementia_variable_start_time
  , a.birth_datetime
  , a.gender_source_value
  , a.ethnicity_concept_id
  , a.death_datetime
  , c.Lsoa
  , d.Index_of_Multiple_Deprivation_Decile
  , d.Income_Decile
  , d.Employment_Decile
  , d.Education_and_Skills_Decile
  , d.Barriers_to_Housing_and_Services_Decile
  , d.Living_Environment_Decile
  from `yhcr-prd-phm-bia-core.CB_FDM_PrimaryCare_V8.person` a
Inner join `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.2024_dementia_2` b
on a.person_id  = b.person_id
left join `yhcr-prd-phm-bia-core.CB_LOOKUPS.tbl_person_lsoa` as c 
on a.person_id = c.person_id
left join `yhcr-prd-phm-bia-core.CB_LOOKUPS.tbl_IMD_by_LSOA` as d 
on c.lsoa = d.LSOA_code
                                                                                      
# clean                                     
CREATE TABLE `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.pw_table1_dementia` as 
SELECT a.*
, b.gender_variable_value
, b.gender_variable_label
, c.ethnicity_variable_value_original
, c.ethnicity_variable_value_11levels
, c.ethnicity_variable_label_11levels
, c.ethnicity_variable_value_8levels
, c.ethnicity_variable_label_8levels
, c.ethnicity_variable_value_5levels
, c.ethnicity_variable_label_5levels 
FROM `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.dementia_3` as a 
join `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.clean_gender` as b on a.gender_source_value = b.gender_source_value 
join `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.clean_ethnic_1220` as c on a.ethnicity_concept_id=c.concept_id                                                                                      
                                                                                                                                                                                                                               
UPDATE `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.pool_over45`
UPDATE `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.pw_table1_dementia`
SET ethnicity_concept_id = recode_ethnicity_concept_id
FROM `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.pw_table1_dementia` a Left Join  `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.impute_ethnicity_1220` b
ON a.person_id=b.person_id
WHERE a.ethnicity_concept_id=8552

#total table
```

```r
CREATE TABLE `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.0202_dementia_1`
AS SELECT distinct person_id, min(tbl_srcode_start_date) dementia_variable_start_time
from `yhcr-prd-phm-bia-core.CB_FDM_PrimaryCare_V9.tbl_srcode` 
WHERE snomedcode in 
(SELECT snomedcode from `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.cb_codelist_dementia_clear`)
GROUP BY person_id 


CREATE TABLE `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.0202_dementia_2` AS
SELECT distinct person_id, snomedcode, min(tbl_srcode_start_date) dementia_variable_start_time, max (numericvalue) reading
from `yhcr-prd-phm-bia-core.CB_FDM_PrimaryCare_V9.tbl_srcode` 
WHERE snomedcode ="408492009"   
GROUP BY person_id, snomedcode

DELETE FROM `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.0202_dementia_2` 
WHERE READING<"8"

ALTER TABLE `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.0202_dementia_2` 
DROP COLUMN snomedcode

ALTER TABLE `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.0202_dementia_2` 
DROP COLUMN reading

CREATE TABLE `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.0202_dementia_all` as
SELECT * FROM `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.0202_dementia_1`  
  UNION ALL 
  SELECT * 
  FROM`yhcr-prd-phm-bia-core.CB_MYSPACE_LG.0202_dementia_2`

ALTER TABLE `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.0202_dementia_all`
ADD COLUMN dementia_variable_value INT

UPDATE `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.0202_dementia_all`
SET dementia_variable_value = 1
WHERE dementia_variable_value IS NULL

CREATE TABLE `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.0202_dementia_all_no_repete` as
SELECT person_id, dementia_variable_value, min(dementia_variable_start_time) dementia_variable_start_time
FROM `yhcr-prd-phm-bia-core.CB_MYSPACE_LG.0202_dementia_all`  
GROUP BY person_id, dementia_variable_value
```
