```r
# create ethnicity table from srpatient (* 947516)
CREATE TABLE `yhcr-prd-phm-bia-core.CB_1322.A_Lin_ethnicity_prep`  AS
SELECT distinct person_id, ethnicity,
  FROM `yhcr-prd-phm-bia-core.CB_FDM_PrimaryCare.tbl_srpatient`
WHERE ethnicity !="Ethnic category - 2011 census England and Wales"  AND
ethnicity !="O/E - ethnic group NOS"  AND
ethnicity !="O/E - ethnic group"  AND
ethnicity !="Ethnic groups (census) NOS"  AND
ethnicity !="Not specified - JDB" 


# delete the repeated rows (* 861620)
CREATE TABLE `yhcr-prd-phm-bia-core.CB_1322.A_Lin_ethnicity`  AS
SELECT person_id , ethnicity,
FROM (
    SELECT ROW_NUMBER() OVER (PARTITION BY person_id ORDER BY RAND()) AS row_num,
        person_id, ethnicity
    FROM  `yhcr-prd-phm-bia-core.CB_1322.A_Lin_ethnicity_prep`
) AS T
WHERE row_num = 1
```
