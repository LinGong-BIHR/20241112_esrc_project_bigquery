##### yearly prevalence

```r
# Define the range of years for prevalence calculation (2009 to 2019)
start_year <- 2009
end_year <- 2019
years <- start_year:end_year

# Create an empty data frame to store yearly prevalence results by ethnicity
prevalence_results <- data.frame(year = integer(), ethnicity = character(), 
                                 total = integer(), dementia_cases = integer(), 
                                 prevalence_rate = numeric())

# Loop through each year to calculate prevalence by ethnicity
for (year in years) {
  # Define the start and end of the year
  start_of_year <- as.Date(paste(year, "-01-01", sep = ""))
  end_of_year <- as.Date(paste(year, "-12-31", sep = ""))
  
  # Loop through each ethnicity level (assuming you have an ethnicity variable in your data)
  ethnicity_levels <- unique(data$ethnicity_variable_label_8levels)  # Replace with your ethnicity column name
  
  for (ethnicity in ethnicity_levels) {
    
    # Subset data for the specific ethnicity and the year range
    ethnicity_data <- subset(data, ethnicity_variable_label_8levels == ethnicity & 
                             entry_Date <= end_of_year)  # Using entry_Date instead of entry_date

    # Exclude individuals who died before the end of the year
    ethnicity_data <- subset(ethnicity_data, is.na(death_datetime) | death_datetime > end_of_year)
    
    # Count dementia cases for individuals diagnosed during the year
    dementia_cases_year <- sum(!is.na(ethnicity_data$dementia_variable_start_time) & 
                               ethnicity_data$dementia_variable_start_time >= start_of_year & 
                               ethnicity_data$dementia_variable_start_time <= end_of_year)
    
    # Count the total number of individuals observed during that year and ethnicity
    total_individuals_year <- nrow(ethnicity_data)
    
    # Calculate the prevalence rate for the year (percentage)
    prevalence_rate <- (dementia_cases_year / total_individuals_year) * 100
    
    # Store the results in the data frame
    prevalence_results <- rbind(prevalence_results, data.frame(year = year, 
                                                              ethnicity = ethnicity, 
                                                              total = total_individuals_year, 
                                                              dementia_cases = dementia_cases_year, 
                                                              prevalence_rate = round(prevalence_rate, 2)))
  }
}

# Print the results
print(prevalence_results)

# Save the results to a CSV file
write.csv(prevalence_results, "yearly_prevalence_by_ethnicity.csv", row.names = FALSE)

# Print a message indicating successful download
cat("Data has been downloaded and saved as 'yearly_prevalence_by_ethnicity.csv'.\n")
```

#### adjusted prevalence by age groups

```r
calculate_dementia_prevalence <- function(data, gender_value, age_range, prevalence_date = "2019-12-31") {
  # Convert prevalence_date to Date format
  prevalence_date <- as.Date(prevalence_date)
  
  # Filter the data based on the provided gender and age range
  filtered_df <- subset(data, gender_variable_value == gender_value & age >= age_range[1] & age <= age_range[2])

  # Determine dementia status at the prevalence date
  filtered_df$dementia_at_prevalence <- ifelse(!is.na(filtered_df$dementia_variable_start_time) & 
                                               filtered_df$dementia_variable_start_time <= prevalence_date, 1, 0)

  # Calculate total and dementia cases by ethnicity within the filtered data
  table_total <- table(filtered_df$ethnicity_variable_label_8levels, useNA = "ifany")  # Total by ethnicity
  table_dementia <- table(filtered_df$ethnicity_variable_label_8levels[filtered_df$dementia_at_prevalence == 1], useNA = "ifany")  # Dementia cases by ethnicity

  # Align dimensions by converting to data frames and merging
  table_total_df <- as.data.frame(table_total)
  table_dementia_df <- as.data.frame(table_dementia)

  # Merge total and dementia data, ensuring all levels are represented
  table_combined <- merge(table_total_df, table_dementia_df, by = "Var1", all = TRUE)
  colnames(table_combined) <- c("ethnicity", "total", "dementia_cases")

  # Replace NAs with 0 for dementia cases where no cases were found
  table_combined$dementia_cases[is.na(table_combined$dementia_cases)] <- 0

  # Calculate prevalence rate as a percentage
  table_combined$prevalence_rate_percent <- (table_combined$dementia_cases / table_combined$total) * 100

  # Append "%" symbol to the prevalence rates
  table_combined$prevalence_rate_percent <- paste0(round(table_combined$prevalence_rate_percent, 2), "%")

  # Return the results
  return(table_combined[, c("ethnicity", "prevalence_rate_percent")])
}
```

crude irr

```r
# Assuming 'ethnicity_variable_label_8levels' includes 'White' as one of the groups.
# Convert ethnicity to a factor variable, ensuring that "White" is the reference category
df$ethnicity_variable_label_8levels <- factor(df$ethnicity_variable_value_8levels,
                                              levels = c("1", "2", "3", "4", "5", "6", "7", "8"))

# Fit a Poisson regression model with ethnicity as the predictor
fit_irr <- glm(dementia ~ ethnicity_variable_label_8levels + offset(log(time_to_dementia)),
               family = "poisson", data = df)

# Calculate IRRs and 95% CIs
irrs <- exp(coef(fit_irr))                    # IRRs
ci <- exp(confint(fit_irr))                   # Confidence intervals for IRRs

# Combine IRRs and CIs into a data frame for reporting
results <- data.frame(
  Ethnicity = names(irrs),
  IRR = round(irrs, 2),
  CI_lower = round(ci[, 1], 2),
  CI_upper = round(ci[, 2], 2),
  Report_CI = paste0(round(irrs, 2), " (", round(ci[, 1], 2), ", ", round(ci[, 2], 2), ")")
)

# Print the results
print(results)

# Export the results to a CSV file
write.csv(results, "1_IRR_crude.csv", row.names = FALSE)
```

#### regression on time to death

```r
# Assuming `dementia_variable_value` is the dementia diagnosis indicator
df_dementia <- subset(df, dementia_variable_value == 1 & !is.na(death_datetime))

library(dplyr)
library(survival)

# Calculate 'time_to_death' in years
df_dementia <- df_dementia %>%
  mutate(
    time_to_death = as.numeric(difftime(death_datetime, dementia_variable_start_time, units = "days")) / 365.25,
    death_status = ifelse(!is.na(death_datetime), 1, 0)  # 1 if death date exists, 0 otherwise
  )

# Convert ethnicity to a factor and set "White" as the reference level
df$ethnicity_variable_label_8levels <- relevel(factor(df$ethnicity_variable_label_8levels), ref = "White")
```

```r
library(survival)

# Fit the Cox model (assuming the model is called 'cox_model')

cox_model <- coxph(Surv(time_to_death, death_status) ~ ethnicity_variable_label_8levels + gender_variable_value + imd, data = df_dementia)

# Extract results
summary_cox <- summary(cox_model)
coefficients <- summary_cox$coef  # Extract coefficients and exp(coef)
conf_int <- summary_cox$conf.int  # Extract confidence intervals

# Combine coefficients, exp(coef), and confidence intervals in one data frame
results <- data.frame(
  Variable = rownames(coefficients),
  Coefficient = coefficients[, "coef"],
  HR = coefficients[, "exp(coef)"],
  CI_lower = conf_int[, "lower .95"],
  CI_upper = conf_int[, "upper .95"],
  p_value = coefficients[, "Pr(>|z|)"]
)

# Write results to CSV
summary(cox_model)
write.csv(results, "cox_regression_results.csv", row.names = FALSE)
```
